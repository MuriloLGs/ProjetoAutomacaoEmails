```python
import pandas as pd 
import pathlib
import win32com.client as win32
```

```python
emails = pd.read_excel(r'Bases de Dados\Emails.xlsx')
lojas = pd.read_csv(r'Bases de Dados\Lojas.csv', sep=';', encoding='latin1')
vendas = pd.read_excel(r'Bases de Dados\Vendas.xlsx')

display(emails.head())
display(lojas.head())
display(vendas.head())
```

```python
vendas = vendas.merge(lojas, on='ID Loja')
display(vendas)
```

```python
dic_lojas = {}

for loja in lojas['Loja']:
    dic_lojas[loja] = vendas.loc[vendas['Loja']== loja, :]
#Exemplo 
display(dic_lojas['Iguatemi Campinas'])
```

```python
indicador_do_dia = vendas['Data'].max()
print(indicador_do_dia)
print(f'{indicador_do_dia.day}/{indicador_do_dia.month}')
```

```python
#Criar e salvar pastas de backup das lojas
backup = pathlib.Path(r'Backup Arquivos Lojas')

arquivos_backup = backup.iterdir()
list_arquivos_backup = [arquivo.name for arquivo in arquivos_backup]

for loja in dic_lojas:
    if loja not in list_arquivos_backup:
        nova_pasta = backup / loja
        nova_pasta.mkdir()
    
    nome_arquivo = (f'{indicador_do_dia.month}_{indicador_do_dia.day}_{loja}.xlsx') 
    local_do_arquivo = backup / loja / nome_arquivo
    dic_lojas[loja].to_excel(local_do_arquivo)
        
```

```python
#Metas anuais e diárias 
meta_faturamento_diaria = 1000
meta_faturamento_anual = 165000
meta_qtdeprodutos_diaria = 4
meta_qtdeprodutos_anual = 120
meta_ticketmedio_diario = 500
meta_ticketmedio_anual = 500
```

```python
for loja in dic_lojas:
  
  vendas_loja = dic_lojas[loja]
  vendas_diarias_loja = vendas_loja.loc[vendas_loja['Data']==indicador_do_dia, :]

  #Calculando os faturamentos
  faturamento_anual = vendas_loja['Valor Final'].sum()
  faturamento_diario = vendas_diarias_loja['Valor Final'].sum()

  #Calculando a diversidade de produtos
  qtde_produtos_anual = len(vendas_loja['Produto'].unique())
  qtde_produtos_diario = len(vendas_diarias_loja['Produto'].unique())

  #Calculando o ticket médio 
  valor_venda = vendas_loja.groupby('Código Venda').sum(numeric_only=True)
  ticket_medio_anual = valor_venda['Valor Final'].mean()
  valor_venda_diaria = vendas_diarias_loja.groupby('Código Venda').sum(numeric_only=True)
  ticket_medio_diario = valor_venda_diaria['Valor Final'].mean()

  #Criando o envio do e-mail e das mensagens que serão enviadas para o e-mail de cada pessoa.

  outlook = win32.Dispatch('outlook.application')

  nome =emails.loc[emails['Loja']==loja,'Gerente'].values[0]
  mail = outlook.CreateItem(0)
  mail.To = emails.loc[emails['Loja']==loja, 'E-mail'].values[0]
  mail.Subject = f'OnePage Dia {indicador_do_dia.day}/{indicador_do_dia.month} - Loja {loja}'

  if faturamento_diario >= meta_faturamento_diaria:
    cor_fat_diario = 'green'
  else: 
    cor_fat_diario = 'red'
  if faturamento_anual >= meta_faturamento_anual:  
    cor_fat_anual = 'green'
  else:
    cor_fat_anual = 'red'
  if qtde_produtos_diario >= meta_qtdeprodutos_diaria:
    cor_qtde_diaria = 'green'
  else: 
    cor_qtde_diaria = 'red'
  if qtde_produtos_anual >= meta_qtdeprodutos_anual:
    cor_qtde_anual = 'green'
  else: 
    cor_qtde_anual = 'red'
  if ticket_medio_diario >= meta_ticketmedio_diario:
    cor_ticket_diario = 'green'
  else: 
    cor_ticket_diario = 'red'
  if ticket_medio_anual >= meta_ticketmedio_anual:
    cor_ticket_anual = 'green'
  else:
    cor_ticket_anual = 'red'

  mail.HTMLBody = f''' 
  <p> Bom dia, {nome}</p>

  <p> O Resultado de ontem <strong>({indicador_do_dia.day}/{indicador_do_dia.month})</strong> da <strong>Loja {loja}</strong>  foi:</p> 

  <table>
        <tr>
          <th>Indicador</th>
          <th>Valor Diário</th>
          <th>Meta Diária</th>
          <th>Cenário Diário</th>
        </tr>
        <tr>
          <td>Faturamento</td>
          <td style="text-align: center">R${faturamento_diario:.2f}</td>
          <td style="text-align: center">R${meta_faturamento_diaria:.2f}</td>
          <td style="text-align: center"><font color="{cor_fat_diario}">◙</font></td>
        </tr>
        <tr>
          <td>Diversidade de Produtos</td>
          <td style="text-align: center">{cor_qtde_diaria}</td>
          <td style="text-align: center">{meta_qtdeprodutos_diaria}</td>
          <td style="text-align: center"><font color="{cor_qtde_diaria}">◙</font></td>
        </tr>
        <tr>
          <td>Ticket Médio</td>
          <td style="text-align: center">R${ticket_medio_diario:.2f}</td>
          <td style="text-align: center">R${meta_ticketmedio_diario:.2f}</td>
          <td style="text-align: center"><font color="{cor_ticket_diario}">◙</font></td>
        </tr>
      </table>
      <br>
      <table>
        <tr>
          <th>Indicador</th>
          <th>Valor Anual</th>
          <th>Meta Anual</th>
          <th>Cenário Anual</th>
        </tr>
        <tr>
          <td>Faturamento</td>
          <td style="text-align: center">R${faturamento_anual:.2f}</td>
          <td style="text-align: center">R${meta_faturamento_anual:.2f}</td>
          <td style="text-align: center"><font color="{cor_fat_anual}">◙</font></td>
        </tr>
        <tr>
          <td>Diversidade de Produtos</td>
          <td style="text-align: center">{qtde_produtos_anual}</td>
          <td style="text-align: center">{meta_qtdeprodutos_anual}</td>
          <td style="text-align: center"><font color="{cor_qtde_anual}">◙</font></td>
        </tr>
        <tr>
          <td>Ticket Médio</td>
          <td style="text-align: center">R${ticket_medio_anual:.2f}</td>
          <td style="text-align: center">R${meta_ticketmedio_anual:.2f}</td>
          <td style="text-align: center"><font color="{cor_ticket_anual}">◙</font></td>
        </tr>
      </table>
          

  <p> Segue em anexo a planilha com todos os dados para mais detalhes.</p> 

  <p> Qualquer duvida estou à disposição.</p>
  <p> Att., Murilo</p>  
  '''

  attachment  = pathlib.Path.cwd() / backup / loja / f'{indicador_do_dia.month}_{indicador_do_dia.day}_{loja}.xlsx'
  mail.Attachments.Add(str(attachment))

  mail.Send()
  print(f'E-mail da Loja {loja} enviado.')


```

```python
faturamento_lojas = vendas.groupby('Loja')[['Valor Final']].sum()
faturamento_lojas_anual= faturamento_lojas.sort_values(by='Valor Final', ascending=False)
display(faturamento_lojas_anual)

nome_arquivo = (f'{indicador_do_dia.month}_{indicador_do_dia.day}_Ranking Anual.xlsx')
faturamento_lojas_anual.to_excel(rf'Backup Arquivos Lojas\{nome_arquivo}')

vendas_diarias = vendas.loc[vendas['Data']==indicador_do_dia, :]
faturamento_lojas_diario = vendas_diarias.groupby('Loja')[['Valor Final']].sum()
faturamento_lojas_diario= faturamento_lojas_diario.sort_values(by='Valor Final', ascending=False)
display(faturamento_lojas_diario)

nome_arquivo = (f'{indicador_do_dia.month}_{indicador_do_dia.day}_Ranking Diário.xlsx')
faturamento_lojas_diario.to_excel(rf'Backup Arquivos Lojas\{nome_arquivo}')
```

```python
outlook = win32.Dispatch('outlook.application')

mail = outlook.CreateItem(0)
mail.To = emails.loc[emails['Loja']=='Diretoria', 'E-mail'].values[0]
mail.Subject = f'Ranking Dia {indicador_do_dia.day}/{indicador_do_dia.month}'
mail.Body = f''' 
Prezados, bom dia

Melhor Loja em faturamento Diário: Loja{faturamento_lojas_diario.index[0]} com faturamento R${faturamento_lojas_diario.iloc[0,0]:.2f}
Pior Loja em fatuamento Diário: Loja {faturamento_lojas_diario.index[-1]} com faturamento R${faturamento_lojas_diario.iloc[-1,0]:.2f}

Melhor Loja em faturamento Anual: Loja{faturamento_lojas_anual.index[0]} com faturamento R${faturamento_lojas_anual.iloc[0,0]:.2f}
Pior Loja em fatuamento Anual: Loja {faturamento_lojas_anual.index[-1]} com faturamento R${faturamento_lojas_anual.iloc[-1,0]:.2f}

Segue em anexo os rankings do ano e do dia de todas as lojas.

Qualquer dúvida estou á disposição.
Att.,
Murilo
'''

attachment  = pathlib.Path.cwd() / backup / f'{indicador_do_dia.month}_{indicador_do_dia.day}_Ranking Anual.xlsx'
mail.Attachments.Add(str(attachment))
attachment  = pathlib.Path.cwd() / backup / f'{indicador_do_dia.month}_{indicador_do_dia.day}_Ranking Diário.xlsx'
mail.Attachments.Add(str(attachment))


mail.Send()
print(f'E-mail da Diretoria enviado.')

```
